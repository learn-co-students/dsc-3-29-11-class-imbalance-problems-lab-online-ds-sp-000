## Predicting Credit Card Fraud
Load the **creditcard.csv.gz** file and preview the data. To load a compressed csv use the optional parameter compression='gzip' within pandas read_csv method as in: `pd.read_csv(filename, compression='gzip')`.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df = pd.read_csv('creditcard.csv.gz', compression='gzip')
df.head()
```

```python
df.shape
```

## Preview the class imbalance

You should see that the dataset has 31 columns. The first is a time field, followed by V1-V28; manual feature engineering done on the backend that we have little information about. Finally, we have the amount of the purchase and a binary Class flag. This last column, Class, is the indication of whether or not the purchase was fradulent, and is the feature we are attempting to predict.

Take a look at how imbalanced this dataset is.

```python
for i in range(2):
    print(df['Class'].value_counts(normalize=(i==1)))
```

## Define the Problem

Define X and y and perform a standard train test split.

```python
X = df.drop(columns=['Class'])
y = df.Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

## Create an initial model

As a baseline, fit a cookie cutter out of the box logistic regression model. Then plot the ROC curve and print out the AUC. We'll use this as a comparison for how our future models perform.

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
import warnings
warnings.simplefilter('ignore')

logreg = LogisticRegression(fit_intercept=False, C=1e12)
log_model = logreg.fit(X_train, y_train)
y_hat_test = log_model.predict(X_test)
y_hat_train = log_model.predict(X_train)

y_score = log_model.decision_function(X_test)
rfp, rtp, thresholds = roc_curve(y_test, y_score)
print(f'AUC Score: {auc(rfp, rtp):.2f}')

ticks = [i/20 for i in range(21)]
plt.style.use('ggplot')
plt.figure(figsize=(10, 8))
plt.plot(rfp, rtp, label='ROC Curve')
plt.plot(np.arange(0, 1, .01), np.arange(0, 1, .01), ls='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.xticks(ticks)
plt.yticks(ticks)
plt.show()
```

```python
from imblearn.over_sampling import SMOTE, ADASYN

weights = [None, 'balanced', {1:2, 0:1}, {1:10, 0:1}, {1:100, 0:1}, {1:1000, 0:1}]
names = ['None', 'Balanced', '2 to 1', '10 to 1', '100 to 1', '1000 to 1']

plt.figure(figsize=(10, 8))

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X, y, test_size=.2, random_state=0)
X_resampled, y_resampled = SMOTE().fit_sample(Xr_train, yr_train)

logregr = LogisticRegression(fit_intercept=False, C=1e12)
log_modelr = logregr.fit(X_resampled, y_resampled)
yr_score = log_modelr.decision_function(Xr_test)
fpr, tpr, thresholdr = roc_curve(yr_test, yr_score)
print(f'AUC Score SMOTE: {auc(fpr, tpr):.4f}')
plt.plot(fpr, tpr, label='SMOTE')

for i in range(len(weights)):
    logreg = LogisticRegression(fit_intercept=False, C=1e12, class_weight=weights[i])
    log_model = logreg.fit(X_train, y_train)
#     y_hat_test = log_model.predict(X_test)
#     y_hat_train = log_model.predict(X_train)

    y_score = log_model.decision_function(X_test)
    rfp, rtp, thresholds = roc_curve(y_test, y_score)
    print(f'AUC Score ({names[i]}): {auc(rfp, rtp):.4f}')
    
    plt.plot(rfp, rtp, label=names[i])
    
plt.plot(np.arange(0, 1, .01), np.arange(0, 1, .01), ls='--', c='b')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.title('ROC Curve')
plt.xticks(ticks)
plt.yticks(ticks)
plt.show()
```
